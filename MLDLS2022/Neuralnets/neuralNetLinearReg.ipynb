### Comparing normal regrassion and the one with neural network.
A neural network can also be a linear regressor; if all hidden layers and activation functions are removed, it is still essentially a neural network, although the simplest conceivable NN. In this situation, it would just be a collection of perceptrons ( minus the activations). So, Neural Networks are more thorough and all-encompassing than simple linear regression, and they can perform as well as (or better than) linear regressions when it comes to nonlinear fitting.
I don't think it's straightforward to argue that model A is superior to model B. In some cases, neural networks perform better than alternative models, such as linear regression. It frequently arises when nonlinearities are present. However, it is critical to analyse before moving on to additional considerations. For the same set of input variables, a linear reg model will have less parameters to estimate than a NN. Then, in order to profit from generalisation and nonlinear mapping, a NN will require a bigger dataset for optimization. So, despite existing nonlinearities, a LINEAR reg model may be better adjusted if we do not have enough data.
The primary distinction in this case (between traditional nonlinear regression and neural network regression) is that in the traditional version, you have already modelled the system with an equation or a system of equations with unknown parameters, which you then fit using regression techniques such as Marquardt-Levenberg. However, with neural networks, you do not have to burden yourself with understanding the underlying physics of the phenomenon and developing a physical model (as you would with traditional nonlinear methodology), you simply set up your network and throw the training inputs at it, and it will minimise the sum of squared errors using gradient descent and backpropagation with minimal intervention from us.
In short, it appears that NN wins in general but in cases linear regression might be better.

```python
#import libs
import tensorflow as tf
import numpy as np
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
from tensorflow import keras
import matplotlib.pyplot as plt
```

```python
#data import+display
train = pd.read_csv('test.csv')
dataset = list(zip(train.x, train.y))
for i in dataset: plt.plot(i[0], i[1], 'g.')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
display(dataset)
```

```python
model = tf.keras.Sequential([keras.layers.Dense(units = 3,activation='relu', input_shape=[1])])
model.add(Dense(1))
print(model.summary())
```

```python
model.compile(optimizer='sgd', loss='mean_squared_error')
```

```python
run_model = model.fit(train.x,train.y, epochs=800)
```

```python
print(model.predict([77]))
```

```python
print(run_model.history.keys())
```

```python
plt.plot(run_model.history['loss'])
plt.title('model loss')
```
